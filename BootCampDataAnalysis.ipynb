## **Import Necessary Libraries**

```python
import plotly.express as px
import pandas as pd
from matplotlib import pyplot as plt
from arabic_reshaper import reshape
from bidi.algorithm import get_display
from matplotlib import pyplot as plt
%matplotlib inline
```

## **Read File as Excel**

```python
df=pd.read_excel(r'vaccination_data_toAnalysis.xlsx')
```

## **Choose An Option**

```python
pd.options.plotting.backend='plotly'
```

## **To see the Negative and Positive as a PieChart by Percentage**

```python
Negative=df.covid_status.value_counts()[0]
Positive=df.covid_status.value_counts()[1]
total=Negative+Positive
percentages=[Negative/total,Positive/total]
_title="نسبة الاصابات"
reshaped=reshape(_title)
transformed=get_display(reshaped)
plt.pie(percentages,labels=['Negative','Positive'],autopct='%.2f%%')
plt.title(transformed)
plt.show()
```

```python
fig=px.pie(data_frame=df,names='covid_status')
fig.show()
```

## **To See The Difference Count Of An Ages As Histogram**

```python
fig=px.histogram(data_frame=df,x='age',title='الفئة العمرية فى الدراسة')
fig.show()
```

## **To See The Difference Count of Country**

```python
fig=px.histogram(data_frame=df,x='region_ar',title='السكان حسب المنطقة')
fig.show()
```
