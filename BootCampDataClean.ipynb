## **Method Helper to Correct_Date** 

```python
def correct_date(date):
    year,month,day = date.split("-")
    if int(year)<1900:
        return pd.to_datetime(Hijri(int(year), int(month), int(day)).to_gregorian())
    else:
        return pd.to_datetime(date)
```

## **Apply it on Date of Birth**

```python
df["DoB"] = df["DoB"].apply(correct_date)
```

## **Convert Date of Birth and VaxDate to Date Type**

```python
df['DoB'] =  pd.to_datetime(df['DoB'])
df['vax_date'] =  pd.to_datetime(df['vax_date'])
```

## **Method Helper to Convert 0.0 to be NotVaccinated and 1.0 to be Vaccinated**

```python
def categorize_status(status):
    if status== '0.0':
        return 'NotVaccinated'
    elif status == '1.0':
        return 'Vaccinated'
    
    else:
        return None
```

## **Method Helper to convert the True to Positive and False to Negative**

```python
def categorize_covid(covid):
    if covid== 'True':
        return 'Positive'
    elif covid == 'False':
        return 'Negative'
    
    else:
        return None
```

## **Apply the two Method  Helper**

```python
df["vax_status"]=df["vax_status"].apply(categorize_status)
```

```python
df["covid_status"]=df["covid_status"].apply(categorize_covid)
```

## **Helper Method to Classify Ages**

```python
def categorize_age(age):
    if age >= 12 and age<=18:
        return "12-18"
    elif age > 18 and age <=30:
        return "18-30"
    elif age > 30 and age < 45:
        return "30-45"
    elif age >= 45:
        return "45>"
    else:
        return "Error"
```

## **Apply Helper Method**

```python
df["age"]=df["age"].apply(categorize_age)
```
